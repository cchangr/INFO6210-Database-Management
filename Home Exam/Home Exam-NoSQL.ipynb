## Mapping from SQL to NoSQL

### The corresponding relationship between SQL database and mongo databse:

* RDBMS -----> MongoDB
* Database -----> Database 
* Table -----> Collection
* Row/Tuple -----> Document 
* Column -----> Field    
* Relationships -----> Embedded Documents / linked file
* Primary Key -----> Primary Key(_id)

### Design description

* Original database: PostgreSQL Database
* Target database: mongodb
* Each databadse in mongodb consist of collections which are equal to sql tables in relational database. 
* each collection stores data with documents form, which is equal to rows/tuples stored in the tables of relational database. 
* in relational databaes we have column, also in mongodb we have field. 
* In order to convert SQL database to mongo database, at first, we need to do denormalization opeartion. Because mongodb do not have JOIN operation, and it stores data using document, we need to denormalize our table and combine them into a big table.
* after denomalization, think about relationship, use aggregate to convert it
* In my sample SQL database, it mainly includes two parts: social media data and games data. social media data has 2 table: game_tweets and game_hashtags. games data has 4 tables: games, games_features, games_reviews, games_users.
* At last It will have 2 collections in mongodb. one is games_data, one is game_twitter. All social media data will be converted to one collection in mongodb, all games data will be converted to one collection in mongodb. 

### Step description

* totally, it includes two parts, loading social meida data and loading domain games data.
* every step of loading two part data has no difference.
* In SQL database, denormalization is first.
* thinking about 1 to n relationship, decide aggregate or embeded file or linked files.
* after a big table generated, export to csv file
* create collections in mongodb
* use python function to import it to mongodb

### 2. load SOCIAL MEDIA data to mongodb.

#### i. denormalization social media data

* use sql command to aggregate 2 table, game_hashtags and game_tweets to one big table.**
* create a view for the new table**

```python
%sql CREATE VIEW full_tweet_data1 AS \
                SELECT B.*, A.hashtags \
                FROM \
                ( \
                    (SELECT hashtags, id FROM game_hashtags) A \
                    FULL OUTER JOIN \
                    (SELECT * FROM game_tweets) B on A.id = B.id);
```

```python
import psycopg2
conn = psycopg2.connect(database='postgres',user='postgres', password='')
```

#### ii. aggregate the 1 to n relationship, export to csv and import to mongodb

**In database it always has some situations that the relationship between table is 1 to n. for example, in my games_users table, 1 game may be have some users. In mongodb these situation should not appear. so in sql I use array_agg function to group 1 to n relationship to 1 row.**

```python
df_k=pd.read_sql('''select id,created_at,tweets,user_id,user_location,retweet_count,"Urls", "Media",game,array_agg(hashtags)
from full_tweet_data1
group by id,created_at,tweets,user_id,user_location,retweet_count,"Urls","Media",game;''', con=conn)
```

```python
df_k
```

**after that, a big table will appear and we will export this table.**

```python
df_k.to_csv('full_tweets_data.csv', encoding='utf-8', index=False)
```

**Then we write a function called mongoimport to import the table we export from RDBMS. this function will read the csv file and insert rows to relational collections of mongodb.**

```python
import pymongo
import pprint
import pandas as pd
from pymongo import MongoClient
import json

def mongoimport(csv_path, db, collection, host='localhost', port=27017):
    client = MongoClient(host, port)
    db = client[db]
    coll = db[collection]
    data = pd.read_csv(csv_path, encoding='latin1')
    load = json.loads(data.to_json(orient='records'))
    coll.remove()
    coll.insert(load)
    return coll.count()
```

```python
db_name='local'
coll_name1='game_twitter'
```

```python
mongoimport(twitter_csv_path, db_name, coll_name1)
```

### 3. load GAMES data to mongodb

#### i. Denormalization GAMES information

**It contains 4 table, use SQL syntax to denormalization to 1 big table**

```python
%sql CREATE VIEW full_games_data1 AS \
SELECT A."Name", B.*, C."AboutText",C."ShortDescrip", C."Reviews", D.user_id \
FROM \
  ( \
    (SELECT "ID","Name" FROM games) A \
    FULL OUTER JOIN \
    (SELECT * FROM games_features) B on A."ID" = B."ResponseID" \
    FULL OUTER JOIN \
    (SELECT * FROM games_reviews) C on A."ID" = C."ResponseID" \
    FULL OUTER JOIN \
    (SELECT * FROM games_users) D on A."ID" = D."ID" \
  )
```
