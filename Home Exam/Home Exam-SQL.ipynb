### 12.Write a function to calculate something relevant to your project. Show that it works.

```python
#the function given games id and return the people critic count.(=Metacritic*0.95+RecommendationCount*0.05)
%sql CREATE OR REPLACE FUNCTION total_feedback(id bigint) \
     RETURNS numeric as \
     $$ \
        SELECT games_features."Metacritic" * 0.95+ games_features."RecommendationCount" * 0.05 FROM games_features where id = "ResponseID"; \
     $$ \
     LANGUAGE SQL;
```

### 26.Create a trigger.

```python
#create a table, two columns emp_id and entry_date.
%sql CREATE TABLE AUDIT(EMP_ID INT NOT NULL, ENTRY_DATE TEXT NOT NULL);
```

```python
#create a function, logfunc.
%sql CREATE OR REPLACE FUNCTION logfunc() RETURNS TRIGGER AS $body$ \
        BEGIN \
            INSERT INTO AUDIT(EMP_ID, ENTRY_DATE) VALUES (new."ID", current_timestamp); \
            RETURN NEW; \
        END; \
    $body$ LANGUAGE plpgsql;
```

```python
#create trigger when happen insert then execute logfunc function.
%sql CREATE TRIGGER trigger111 AFTER INSERT ON games \
     FOR EACH ROW EXECUTE PROCEDURE logfunc();
```

### 27.Create a transaction.

```python
import pandas as pd
conn = psycopg2.connect("dbname=postgres user=postgres")
curs = conn.cursor()
```

```python
curs.execute('''BEGIN TRANSACTION; 
     SELECT * FROM games WHERE "ID"=20000; 
     UPDATE games SET "Name"='test100' WHERE "ID"=20000; 
    ROLLBACK;''')
```

```python
%sql select * from games where "ID"=20000
```

```python
curs.execute('''BEGIN TRANSACTION; 
     SELECT * FROM games WHERE "ID"=20000; 
     UPDATE games SET "Name"='test100' WHERE "ID"=20000; 
    COMMIT;''')
```

### 28.Explain to an eight year old (i.e. your professor) what are the first three Normal Forms. (10 points)

##### First Normal Form :
* at first it should have single valued columns, every field should be single
* values in the same column should belong to the same domain.
* every column  should have a unique name in table.
* the data order is unimportant.

##### Second Normal Form :
* at first belongs to first normal form
* no functional dependency

##### Third Normal Form :
* at first belongs to Second Normal Form
* non-primany key should have no relationship between each other.
