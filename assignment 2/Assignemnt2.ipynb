```python
#Getting Tweets
import json

search_words = ['#WiiSports','#Pokemon','#SuperMarioBro','#gtaV','#DarkSouls3','#Halo','#finalfantasy','#rayman',
         '#FIFA16','#Fallout']
search_games = ['Wii Sports','Pokemon Red/Pokemon Blue','Super Mario Bros.','Grand Theft Auto V','Dark Souls III',
         'Halo Wars','Final Fantasy V','Rayman','FIFA 16','Fallout 4']

q = 'playstation' 
n = 1050
from urllib.parse import unquote
search_results = api.search(q = q, count = n)
```

```python
#connect the database and write data to sql
USER = "postgres"
PASSWORD = ""
DATABASE = "postgres"
```

```python
def create_twitter_table():
    conn = psycopg2.connect(database=DATABASE,user=USER, password=PASSWORD)
    cur = conn.cursor()
    cur.execute('''CREATE TABLE game_twitter(screen_name varchar,created_at varchar,text varchar)''')
    conn.commit()
    conn.close()
```

```python
#write the tweets from twitter to postgresql database
def write_tweets_sql(created_at, text, screen_name):
    db= psycopg2.connect(database=DATABASE, user=USER, password=PASSWORD)
    cursor = db.cursor()
    insert_query = "INSERT INTO game_twitter(screen_name,created_at,text) VALUES ( %s, %s, %s)"
    cursor.execute(insert_query, (screen_name, created_at, text))
    db.commit()
    cursor.close()
    db.close()
    return
def store_tweets():
    for word in search_words:
        for tweet in api.search(q = word, count = 100):
            write_tweets_sql(tweet.created_at,tweet.text,tweet.user.screen_name)
            print(tweet.created_at,tweet.text, tweet.user.screen_name)
```

```python
# for every game create one table about their tweets
create_twitter_table()
```

```python
store_tweets()
```

```python
def create_hashtag_table():
    conn = psycopg2.connect(database=DATABASE,user=USER, password=PASSWORD)
    cur = conn.cursor()
    cur.execute('''CREATE TABLE game_tags(tag varchar, game varchar)''')
    conn.commit()
    conn.close()
```

```python
def write_tags_sql(tag,game):
    db= psycopg2.connect(database=DATABASE, user=USER, password=PASSWORD)
    cursor = db.cursor()
    insert_query = "INSERT INTO game_tags(tag,game) VALUES ( %s, %s)"
    cursor.execute(insert_query, (tag,game))
    db.commit()
    cursor.close()
    db.close()
    return
```

```python
def store_hashtag():
    for i in range (0,10):
        word=search_words[i];
        game=search_games[i];
        for tweet in api.search(q = word, count = 100):
            for hashtag in tweet.entities.get('hashtags'):
                write_tags_sql(hashtag.get('text'),game)
                print(hashtag, game)
```

```python
user1 = api.get_user('Justin_G_Bryan')
print(user1.name, user1.screen_name, str(user1.statuses_count), str(user1.friends_count), str(user1.followers_count))
```

```python
from datetime import datetime, date, time, timedelta
tweets = user1.statuses_count
account_created_date = user1.created_at
delta = datetime.utcnow() - account_created_date
account_age_days = delta.days
print("Account age (in days): " + str(account_age_days))
if account_age_days > 0:
    print("Average tweets per day: " + "%.2f"%(float(tweets)/float(account_age_days)))
```

```python
from tweepy import Cursor
hashtags = []
mentions = []
tweet_count = 0
end_date = datetime.utcnow() - timedelta(days=30)
for status in Cursor(api.user_timeline, id='Justin_G_Bryan').items():
      tweet_count += 1
      if hasattr(status, "entities"):
        entities = status.entities
        if "hashtags" in entities:
          for ent in entities["hashtags"]:
            if ent is not None:
              if "text" in ent:
                hashtag = ent["text"]
                if hashtag is not None:
                  hashtags.append(hashtag)
        if "user_mentions" in entities:
          for ent in entities["user_mentions"]:
            if ent is not None:
              if "screen_name" in ent:
                name = ent["screen_name"]
                if name is not None:
                  mentions.append(name)
      if status.created_at < end_date:
        break
```
