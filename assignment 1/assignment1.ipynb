```python
#SQL USE CASES 1--5
#use case 1
df_r=pd.read_sql('''SELECT PCgames_attribute.Platform, PCgames_attribute.genre, PCgames_sales.NA_sales,
PCgames_sales.JP_sales FROM PCgames_attribute join PCgames_sales
ON PCgames_attribute.Name = PCgames_sales.Name
''', con=conn)
```

```python
df_r.head()
```

```python
#use case 2
df_s=pd.read_sql('''SELECT PCgames_attribute.Name, PCgames_attribute.year_of_release, PCgames_attribute.developer,
PCgames_sales.EU_sales FROM PCgames_attribute JOIN PCgames_sales ON PCgames_attribute.Name = PCgames_sales.Name
ORDER BY PCgames_sales.EU_sales DESC''', con=conn)
```

```python
df_s.head()
```

```python
#use case 3
df_t=pd.read_sql('''SELECT PCgames_attribute.Name, PCgames_attribute.genre, PCgames_sales.Global_Sales
FROM PCgames_attribute INNER JOIN PCgames_sales ON PCgames_attribute.Name = PCgames_sales.Name
WHERE PCgames_sales.Global_Sales > 20 AND PCgames_attribute.genre = 'Sports'
ORDER BY PCgames_sales.Global_Sales DESC''', con=conn)
```

```python
df_t.head()
```

```python
#use case 4
df_v=pd.read_sql('''SELECT PCgames_attribute.Platform, sum(PCgames_sales.NA_sales) as num 
FROM PCgames_attribute INNER JOIN PCgames_sales ON PCgames_attribute.Name = PCgames_sales.Name 
GROUP BY PCgames_attribute.Platform ORDER BY num DESC LIMIT 10''', con = conn)
```

```python
df_v
```

```python
#use case 5
df_w=pd.read_sql('''SELECT PCgames_attribute.Name, PCgames_sales.NA_sales, PCgames_sales.EU_sales,
PCgames_sales.EU_sales, PCgames_sales.JP_sales, PCgames_sales.Other_sales FROM 
PCgames_attribute INNER JOIN PCgames_sales ON PCgames_attribute.Name = PCgames_sales.Name
WHERE PCgames_attribute.Platform = 'X360' ORDER BY PCgames_sales.EU_sales DESC LIMIT 20''', con=conn)
```
